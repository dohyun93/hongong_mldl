```python
# 패션 MNIST 데이터 불러오기
from tensorflow import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) =\
keras.datasets.fashion_mnist.load_data()

# (48000, 28, 28) -> (48000, 28, 28, 1) 크기로 변환하고 255로 나눈 표준화 전처리.
train_scaled = train_input.reshape(-1, 28, 28, 1) / 255.0
train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, random_state=42, test_size=0.2)
```

```python
# 합성곱 신경망 만들기
# keras.Sequential() 로 입력층 이후의 각 레이어를 순차적으로 정의해보자.

#1. 합성곱 층 (1)
# 필터 개수: 32개
# 커널 크기: (3, 3)
# 스트라이드: 1 (=default)
# 패딩: 세임 패딩 (!=default)
model = keras.Sequential()
model.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=(28, 28, 1), strides=1))
```

```python
#2. 풀링 층 (1)
# 전형적인 풀링 크기인 (2, 2)로 맥스풀링 층 추가.
model.add(keras.layers.MaxPooling2D(2))
```

```python
# 합성곱 층에서 세임패딩을 거쳤기 때문에, (28, 28, 32) 가 될 것이고, 
# 풀링 층에서 (2, 2) 크기의 풀링을 거쳤기 때문에, 최종 피쳐맵의 shape은 (14, 14, 32)가 된다.

# 합성곱 층 (2)
# 필터 개수: 64개
# 커널 크기: (3, 3)
# 스트라이드: 1 (=Default)
# 패딩: 세임 패딩 (!=default)
model.add(keras.layers.Conv2D(64, padding='same', strides=1, kernel_size=(3,3), activation='relu'))

# 풀링 층 (2)
# 풀링 크기: (2, 2)
model.add(keras.layers.MaxPooling2D(2))
```

```python
# 두 번째 합성곱 층을 거치면서 (14, 14, 64)의 피쳐맵이 생성되며, 두 번째 풀링 층을 거치면서 (7, 7, 64) 피쳐맵이 만들어진다.

# 이제 최종적으로 이 3차원 피쳐맵을 일렬로 펼쳐서 밀집층을 통과하게 해보자.
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))

# 출력층과 밀집층 사이에 드롭아웃층을 넣어서 모델이 과대적합 되는 것을 방지해주자.
model.add(keras.layers.Dropout(0.4))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
# 이제, 만들어진 모델의 개요를 확인해보자.
print(model.summary())
```

```python
# 이렇게 model.summary() 로도 만든 모델의 전체적인 정보를 알 수 있지만, 도식화 해주는 방법도 있다.
# keras.utils 에 있는 plot_model() 함수를 사용해보자

keras.utils.plot_model(model)
```

```python
# 세부적으로 아래 매개변수를 주어 더 자세한 도식을 확인할 수 있다.
# show_shapes (True/False) : 입력과 출력의 크기를 표현
# dpi: 해상도
# to_file: 출력한 이미지를 파일로 저장.
keras.utils.plot_model(model, dpi=100, to_file='cnn-architecture.jpg', show_shapes=True)
```

```python
# 모델 컴파일과 훈련
# 먼저 만든 모델을 loss, optimizer, metrics를 지정하자
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy', optimizer='adam')

checkpoint_cb = keras.callbacks.ModelCheckpoint('best-cnn-model.h5')
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

history = model.fit(train_scaled, train_target, epochs=20, validation_data=(val_scaled, val_target), callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
# 손실 그래프를 그려서 조기 종료가 잘 이루어졌는지 확인해보자
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()

# [7] 에포크가 가장 최적으로 생각할 수 있다.
```

```python
# EarlyStopping 클래스에서 restore_best_weights=True 로 설정했으므로, 현재 model 객체가 최적의 모델 파라미터로 복원되어 있다.
# ModelCheckpoint 콜백이 저장한 best-cnn-model.h5 파일을 다시 읽을 필요가 없다.

# 검증 세트에 대한 성능을 평가해보자.
model.evaluate(val_scaled, val_target)
```

```python
# 예상했던대로 8/20 epoch 의 검증 점수를 최적이라고 판단했다.
# predict() 메서드를 활용해서 새로운 데이터에 대해 예측을 만들어보자.

# 검증세트의 첫 번째 샘플을 처음 본 이미지라고 가정하고, 확인해보자.
# 검증세트[1]의 shape이 (28, 28, 1) 이나, matplotlib 은 2D 만 처리가능하므로 reshape 해주자.

plt.imshow(val_scaled[0].reshape(28, 28), cmap='gray_r')
plt.show()
```

```python
# 이 샘플에 대한 예측 결과를 확인해보자.
preds = model.predict(val_scaled[0:1])
print(preds)

# 0:1 로 한 이유는?
# fit(), predict(), evaluate() 메서드들은 모두 입력의 첫 번째 차원이 배치 차원이라고 가정한다.
# 따라서 샘플 하나를 전달할 때, (28, 28, 1)이 아니라, (1, 28, 28, 1) 크기를 전달해야 한다.
# 배열 슬라이싱은 인덱싱과 다르게 선택된 원소가 하나더라도, 전체 차원이 유지되어 (1, 28, 28, 1) 크기를 만든다.
```

# 8-3. 합성곱 신경망의 시각화

```python
# 학습했던 최적의 모델 파라미터를 불러온다.
from tensorflow import keras
model = keras.models.load_model('best-cnn-model.h5')

model.layers
```

```python
# 첫 번째 합성곱 층의 가중치를 조사해보자.
# 층의 가중치와 절편은 층의 weights 속성에 저장되어 있다.
# weights 도 파이썬 리스트다.
conv1 = model.layers[0]
print(conv1.weights[0].shape, conv1.weights[1].shape)
```

```python
# weights 리스트의 첫 번째 원소는 가중치를, 두 번째 원소는 절편을 나타낸다.
# weights 속성은 tensorflow 의 다차원 배열인 Tensor class 의 객체이다.
# 다루기쉽게 np 배열로 변환해보자.

conv1_weights = conv1.weights[0].numpy()
print(conv1_weights.mean(), conv1_weights.std())
```

```python
# 첫 번째 합성곱 층의 가중치 평균은 -0.028 이고, 표준편차는 0.27이다.
# 이 가중치가 어떤 분포를 보이는지 시각화해보자.

import matplotlib.pyplot as plt

plt.hist(conv1_weights.reshape(-1, 1)) # 1차원 배열로 전달
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
# 이번에는 conv1 층의 필터들을 시각화해보자.
fig, axs = plt.subplots(2, 16, figsize=(15, 2))
for i in range(2):
  for j in range(16):
    axs[i, j].imshow(conv1_weights[:, :, 0, i*16 + j], vmin=-0.5, vmax=0.5)
    axs[i, j].axis('off')
plt.show()
```

```python
# 위 필터 이미지의 해석을 해보자면, 밝은 부분이 값이 높다.
# 즉, 밝은 부분에 대한 특성을 잘 잡아내는 필터이다.

# imshow() 함수는 배열에 있는 최대값과 최소값을 사용해 픽셀의 강도를 표현한다.
# 따라서 어떤 절대값을 두고 그걸 기준으로해서 픽셀의 강도를 표현하기 위해, vmin, vmax로 matplotlib의 컬러맵으로 표현할 범위를 지정했다.

# 이번에는 최적이라고 학습한 모델파라미터와, 빈 합성곱 층의 모델 파라미터를 비교하기 위해 훈련하지 않은 빈 합성곱 신경망을 만들어본다.
no_training_model = keras.Sequential()
no_training_model.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=(28, 28, 1)))
no_training_conv = no_training_model.layers[0]

print(no_training_conv.weights[0].shape) # 가중치
```

```python
no_training_weights = no_training_conv.weights[0].numpy()
print(no_training_weights.mean(), no_training_weights.std())
```

```python
plt.hist(no_training_weights.reshape(-1, 1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
# 학습한 모델의 가중치 분포도와다르게 고르게 분포되어있는걸 알 수 있다.
# 이제 가중치들을 시각화해보자.
fig, axs = plt.subplots(2, 16, figsize=(15, 2))
for i in range(2):
  for j in range(16):
    axs[i, j].imshow(no_training_weights[:, :, 0, i*16 + j], vmin=-0.5, vmax=0.5)
    axs[i, j].axis('off')
plt.show()
```

```python
# 전체적으로 가중치가 밋밋하게 초기화 된 것을 알 수 있다.

# 이런 상태에서 위에 살펴본 훈련된 모델의 커널로 학습되면서 fashion mnist 데이터의 분류 정확도를 높이기 위해
# 유용한 패턴을 학습했다는 사실을 알 수 있다.

# 한편, 위 모델은 keras.Sequential() 로 차례로 쌓은 모델로, predict() 메서드를 호출하면 입력부터 마지막층까지 모든 계산을 완료한 뒤
# 최종 출력을 반환한다. conv1 레이어에 대해 입력과 출력을 확인하고 그 층의 필터들을 확인하고 싶을 경우, 층별로 케라스 함수형 API 를 
# 사용하는 것이 더욱 유용하다.

# Sequential()은 자동으로 InputLayer 클래스를 추가해주고 호출해주지만, Model 클래스에서는 수동으로 만들어서 호출해줘야한다.
# 마치 inputs = keras.Input(shape=(784,)) 처럼.

# 1. 모델의 입력층 다음레이어인 conv1층의 인풋: 입력층과 동일하며, 'input'속성에 저장.
print(model.input)
# print(model.layers[0].input) 과 동일.

# 2. 모델의 conv1 아웃풋: 
print(model.layers[0].output)

# 위 둘을 활용해서 conv_acti 라는 합성곱 모델을 만들어보자.
conv_acti = keras.Model(model.input, model.layers[0].output)
```

```python
# 위 conv_acti 를 활용한 피쳐맵 시각화
(train_input, train_target), (test_input, test_target) =\
keras.datasets.fashion_mnist.load_data()

print(train_input[0].shape)
plt.imshow(train_input[0], cmap='gray_r')
plt.show()
```

```python
# 이 샘플을 conv_acti 모델에 주입해서 conv2d 층이 만드는 피쳐맵을 출력해보자.
inputs = train_input[0:1].reshape(-1, 28, 28, 1) / 255.0
features_maps = conv_acti.predict(inputs)

# conv_acti.predict()가 출력한 output인 feature_maps 크기를 확인해보자.
print(features_maps.shape)
```

```python
# same padding, 32개의 필터를 사용하였으므로 (28, 28, 32) 크기의 피쳐맵이 나왔다.
fig, axs = plt.subplots(4, 8, figsize=(15, 8))
for i in range(4):
  for j in range(8):
    axs[i, j].imshow(features_maps[0, :, :, i*8 + j])
    axs[i, j].axis('off')
plt.show()
```

```python
# 이 피쳐맵은 32개의 필터로 인해 입력 이미지에서 강하게 활성화된 부분을 보여준다.
# 앞서 32개 필터의 가중치를 출력한 그림과 위 필터 그림 몇 개를 비교해보자. 

# 12번째 필터는 전체적으로 밝은색이므로, 신발에 대한 모든 영역을 감지한다.
# 또한 마지막 필터를 보면 가로, 세로, 가로에 대한 밝은 부분을 볼 수 있는데, conv1_weights로 살펴본 피쳐맵의 마지막
# 컬러맵을 보면 가로, 세로, 가로에 대해 밝은모습을 띄고있는걸 알 수 있다.

# 두 번째 합성곱 층이 만든 피쳐맵도 같은 방식으로 확인할 수 있다.
conv2_acti = keras.Model(model.input, model.layers[2].output)

inputs = train_input[0:1].reshape(-1, 28, 28, 1) / 255.0
feature_maps = conv2_acti.predict(inputs)

print(feature_maps.shape)
```

```python
fig, axs = plt.subplots(8, 8, figsize=(12, 12))
for i in range(8):
  for j in range(8):
    axs[i, j].imshow(feature_maps[0, :, :, i*8 + j])
    axs[i, j].axis('off')
plt.show()
```

```python
# 두 번째 합성곱층의 피쳐맵은 첫 번째 합성곱 층의 피쳐맵보다는 다소 직관적으로 이해하기 어렵다.
# 두 번째 합성곱층의 필터 크기는 (3, 3, 64) 였다.

# 64개의 두 번째 합성곱층의 각 필터가 첫 번째 합성곱층의 풀링결과 (14, 14, 32) 특성맵과 곱해져 64개의 피쳐맵을 만들고,
# 이는 첫 번째 합성곱층의 출력인 피쳐맵 (14, 14, 32)에서 어떤 부위를 감지하는지 직관적으로 이해하기 어렵기 때문이다.

# 즉, 이런 현상은 합성곱 층을 많이 쌓을 수록 심해지는데, 바꾸어 생각하면 
# (1) 합성곱 신경망의 앞부분에 있는 합성곱 층은 이미지의 시각적인 정보를 감지하고,
# (2) 뒤쪽에 있는 합성곱 층은 앞쪽에서 감지한 시각적인 정보를 바탕으로, 추상적인 정보를 학습한다고 볼 수 있다.
```
